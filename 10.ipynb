1. Using the XOR dataset below, train (400 epochs) a neural network (NN) using 1, 2, 3, 4, and 5 hidden layers (where each layer has only 2 neurons). For each n layers, store the resulting loss score along with n. Plot the results to find what the optimal number of layers is.

4. Instead of `SGD` try other optimizers and report on the loss score. (https://keras.io/optimizers/)

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD  #Stochastic Gradient Descent
import pandas as pd
import numpy as np
# fix random seed for reproducibility
np.random.seed(7)

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 40
xx = np.random.random((n,1))
yy = np.random.random((n,1))
```

```python
X = np.array([np.array([xx,-xx,-xx,xx]),np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
y = np.array([np.ones([2*n]),np.zeros([2*n])]).reshape(4*n)
```

```python
plt.scatter(*zip(*X), c=y)
```

## Question 1

```python
# 1 layer
model_1 = Sequential()
model_1.add(Dense(2, input_dim=2, activation='tanh'))
model_1.add(Dense(1, activation='sigmoid'))
sgd = SGD(lr=0.1)
model_1.compile(loss='binary_crossentropy', optimizer='sgd')
model_1.fit(X, y, batch_size=2, epochs=400) 
scores_1 = model_1.evaluate(X, y)

# 2 layers
model_2 = Sequential()
model_2.add(Dense(2, input_dim=2, activation='tanh'))
model_2.add(Dense(2, input_dim=2, activation='tanh'))
model_2.add(Dense(1, activation='sigmoid'))
sgd = SGD(lr=0.1)
model_2.compile(loss='binary_crossentropy', optimizer='sgd')
model_2.fit(X, y, batch_size=2, epochs=400) 
scores_2 = model_2.evaluate(X, y)

# 3 layers
model_3 = Sequential()
model_3.add(Dense(2, input_dim=2, activation='tanh'))
model_3.add(Dense(2, input_dim=2, activation='tanh'))
model_3.add(Dense(2, input_dim=2, activation='tanh'))
model_3.add(Dense(1, activation='sigmoid'))
sgd = SGD(lr=0.1)
model_3.compile(loss='binary_crossentropy', optimizer='sgd')
model_3.fit(X, y, batch_size=2, epochs=400) 
scores_3 = model_3.evaluate(X, y)

# 4 layers
model_4 = Sequential()
model_4.add(Dense(2, input_dim=2, activation='tanh'))
model_4.add(Dense(2, input_dim=2, activation='tanh'))
model_4.add(Dense(2, input_dim=2, activation='tanh'))
model_4.add(Dense(2, input_dim=2, activation='tanh'))
model_4.add(Dense(1, activation='sigmoid'))
sgd = SGD(lr=0.1)
model_4.compile(loss='binary_crossentropy', optimizer='sgd')
model_4.fit(X, y, batch_size=2, epochs=400) 
scores_4 = model_4.evaluate(X, y)

# 5 layers
model_5 = Sequential()
model_5.add(Dense(2, input_dim=2, activation='tanh'))
model_5.add(Dense(2, input_dim=2, activation='tanh'))
model_5.add(Dense(2, input_dim=2, activation='tanh'))
model_5.add(Dense(2, input_dim=2, activation='tanh'))
model_5.add(Dense(2, input_dim=2, activation='tanh'))
model_5.add(Dense(1, activation='sigmoid'))
sgd = SGD(lr=0.1)
model_5.compile(loss='binary_crossentropy', optimizer='sgd')
model_5.fit(X, y, batch_size=2, epochs=400) 
scores_5 = model_5.evaluate(X, y)

x_axis = range(1,6)
plt.plot(x_axis, [scores_1,scores_2,scores_3,scores_4,scores_5])
plt.xlabel('Number of Layers')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
# load pima indians dataset
import numpy
dataset = numpy.loadtxt("C:\Users\walte/Desktop/mlnn/data/pima-indians-diabetes.data", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
plt.plot(X,Y)
```

```python
# create model
model = Sequential()
model.add(Dense(16, input_dim=8, activation='tanh'))
model.add(Dense(16, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Fit the model
model.fit(X, Y, epochs=1000, batch_size=10)
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
