# Question 1

```python
word = []
for text in processed_text:
    if not text.is_stop and not text.is_punct:
        word.append(text.text)
        
def counts(my_list):
    counts = dict()
    for item in my_list:
        counts[item] = my_list.count(item)
        
    return counts

most = counts(word)
most = dict(sorted(most.items(), key=lambda item: item[1], reverse=True))
most
```

# Question 2

```python
pos = []
for text in processed_text:
    if not text.is_stop and not text.is_punct:
        if text.pos_ not in pos:
            pos.append(text.pos_)
pos
```

```python
most_pos = dict()
for each in pos:
    word_pos = []
    for text in processed_text:
        if not text.is_stop and not text.is_punct and text.pos_==each:
            word_pos.append(text.text)

    most = counts(word_pos)
    most = dict(sorted(most.items(), key=lambda item: item[1], reverse=True))
    most_pos[each] = most
most_pos
```

```python
def pr_tree(word, level):
    if word.is_punct:
        return
    for child in word.lefts:
        pr_tree(child, level+1)
    print('\t' * level + word.text + ' - ' + word.dep_)
    for child in word.rights:
        pr_tree(child, level + 1)
```

```python
nlp = spacy.load('en_core_web_lg')
processed_new = nlp(article)
sim = dict()
for text in processed_new:
    word = []
    if not text.is_stop and not text.is_punct:
        if text.text not in word:
            word.append(text.text)
    for w in word:
        tok = nlp(w)
        sim[tok.text] = dict()
        for token in processed_new:
            if not token.is_stop and not token.is_punct: 
                sim[tok.text][token.text] = (tok.similarity(token))
```

```python
res = {key: dict(list(sorted(val.items(), key = lambda ele: ele[1], reverse=True))[0:5])
      for key, val in sim.items()}
res
```
